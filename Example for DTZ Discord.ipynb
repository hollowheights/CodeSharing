# Load data

```python
# --- IMPORTS AND SIMILAR ---#
import time
import datetime
from dateutil import tz
import sqlite3
import matplotlib.pyplot as plt
import pandas as pd

from IPython.display import display, HTML
display(HTML("<style>.container { width:80% !important; }</style>"))

stock = "ISIG"
startdate = "2017-03-13"
enddate = "2017-03-19"

# -------------- Connect to local SQLite db --------------------
#connect to SQLite database
connect = sqlite3.connect(r"C:\01 SR Data partition\TradingDatabaser\SRUSListedPolygonDaily.db")
c = connect.cursor()
```

```python
df = (pd.read_sql("SELECT * FROM %s WHERE DATE(date) BETWEEN '%s' AND '%s' " % (stock, startdate, enddate), connect, parse_dates=['Date']) )
df.drop(["Weekday"],axis=1, inplace=True) #deleting the Weekday column just to be able to add it below
df
```

# Add a weekday column - days by name

```python
df['Weekday'] = df.Date.dt.day_name()
df
```

# Add a weekday column - days by number (0 = monday)

```python
df['Weekday_numbers'] = df.Date.dt.dayofweek
df
```
